# traffic nature numeric encoding

```python
NON_thor_label = 0
VPN_label = 2
NON_VPN_label = 1
THOR_label =3
```

```python
trafficNature = df['traffic nature']
for i,value in enumerate(trafficNature):
    if value == 'Non-Tor':
        trafficNature[i] = NON_thor_label
    elif value == 'NonVPN':
        trafficNature[i] = NON_VPN_label
    elif value == 'VPN':
        trafficNature[i] = VPN_label
    elif value =='Tor':
        trafficNature[i] = THOR_label
```

```python
trafficNature
```

```python
df['traffic nature'] = trafficNature
```

# IP adresses numeric encoding :

```python
import socket
import struct

ip_address = '192.168.0.1'

packed_ip = socket.inet_aton(ip_address)
ip_int = struct.unpack('!L', packed_ip)[0]

print(ip_int)  
```

```python
def IPtoNumeric(ip):
    packed_ip = socket.inet_aton(ip)
    ip_int = struct.unpack('!L', packed_ip)[0]
    return ip_int
```

```python
srcIP = df["Src IP"]
numericSrcIP = srcIP.map(IPtoNumeric)
dstIP = df['Dst IP']
numericDstIP = dstIP.map(IPtoNumeric)
```

```python
df["Src IP"] = numericSrcIP
df["Dst IP"] = numericDstIP
```
